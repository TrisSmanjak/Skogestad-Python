# Frequency response

```python
def G(s):
    return 1/(s + 1)
```

```python
ω = numpy.logspace(-2, 1)
```

```python
s = 1j*ω
```

```python
plt.loglog(ω, numpy.abs(G(s)**3))
```

```python
plt.semilogx(ω, numpy.unwrap(numpy.angle(G(s)**3)))
```
