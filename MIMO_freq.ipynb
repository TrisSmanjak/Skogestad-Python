```python
import numpy
import scipy.signal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def G(s):
    return numpy.array([[4/(s + 1), 2/(3*s + 1)], 
                       [2/(2*s + 1), 3/(2*s + 1)]])

s = numpy.logspace(-2, 2, 1000)
omega = 1j*s

S = [numpy.linalg.svd(G(w))[1] for w in omega]
sg_u = [s[0] for s in S]
sg_l = [s[1] for s in S]

plt.loglog(s, sg_u)
plt.loglog(s, sg_l)
```

```python
rga = G(0)*numpy.transpose((numpy.linalg.inv(G(0))))
rga
```

```python
rga = G(1j)*numpy.transpose((numpy.linalg.inv(G(1j))))
rga
```

```python
rgas = [G(w)*numpy.transpose((numpy.linalg.inv(G(w))))[0, 0] for w in omega]
rgas = [rg[0][0] for rg in rgas]
plt.semilogx(s, rgas)
```
