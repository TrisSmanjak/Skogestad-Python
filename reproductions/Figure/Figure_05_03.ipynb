```python
import numpy as np
import control
```

```python
def L(s):
    return (1/(s+1))*((-s+1)/(s+1))

def Lm(s):
    return (1/(s+1))
```

```python
w = np.logspace(-2, 2, 1000)

L = control.tf([-1, 1], [1, 2, 1])
re, im, fr = control.nyquist(L, w, Plot=False)

Lm = control.tf([1], [1, 1])
rem, imm, fr = control.nyquist(Lm, w, Plot=False)
```
